## Section 1 — Array Creation

**Drill 1.1** — 3x4 array with random integers between 10 and 50.

```python
arr = np.random.randint(10, 50, (3, 4))
print(arr)
```

**Drill 1.2** — 20 evenly spaced numbers between 0 and 5.

```python
arr = np.linspace(0, 5, 20)
print(arr)
```

---
## Section 2 — Indexing and Slicing

**Drill 2.1** — Extract center 3x3 from a 5x5 matrix.

```python
arr = np.random.randint(10, 100, (5, 5))
center = arr[1:4, 1:4]
print("5x5:\n", arr)
print("center 3x3:\n", center)
```

**Drill 2.2** — Transpose a 2x2 array.

```python
arr = np.random.randint(10, 100, (2, 2))
print("original:\n", arr)
print("transposed:\n", arr.T)
```

---
## Section 3 — Shape Transformations

**Drill 3.1** — Reshape a (2,6) array into (3,4) and (4,3).

```python
arr = np.random.randint(10, 100, (2, 6))
print("(3,4):\n", arr.reshape(3, 4))
print("(4,3):\n", arr.reshape(4, 3))
```

**Drill 3.2** — Flatten a 2D array and reshape back to original shape.

```python
arr = np.random.rand(2, 2)
flat = arr.flatten()
restored = flat.reshape(arr.shape)
print("flat:", flat)
print("restored:\n", restored)
```

---
## Section 4 — Masking and Filtering

**Drill 4.1** — Replace all values less than 0 with 0.

```python
arr = np.random.randint(-10, 50, (4, 4))
arr[arr < 0] = 0
print(arr)
```

**Drill 4.2** — Count values greater than the mean.

```python
arr = np.random.randint(10, 50, (4, 4))
count = np.sum(arr > arr.mean())
print("count > mean:", count)
```

---
## Section 5 — Conditional Replacement

**Drill 5.1** — Replace all even numbers with their negative value.

```python
arr = np.random.randint(10, 50, size=10)
arr[arr % 2 == 0] *= -1
print(arr)
```

---
## Section 6 — Row and Column Operations

**Drill 6.1** — Normalize each row so the row sum equals 1.

```python
arr = np.random.randint(1, 10, (4, 4))
normalized = arr / arr.sum(axis=1, keepdims=True)
print("original:\n", arr)
print("normalized:\n", normalized)
print("row sums:", normalized.sum(axis=1))
```

**Drill 6.2** — Set all diagonal elements to 0.

```python
arr = np.random.randint(1, 10, (4, 4))
np.fill_diagonal(arr, 0)
print(arr)
```
