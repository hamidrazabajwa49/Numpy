# NumPy — ML Applications

20 exercises applying NumPy to real ML preprocessing tasks: missing value imputation, normalization, one-hot encoding, moving averages, matrix multiplication, and more.

```python
import numpy as np

np.random.seed(0)
```

## 1. 3x4 Array — Random Integers (10–50)

```python
arr = np.random.randint(10, 50, (3, 4))
print(arr)
```

## 2. 20 Evenly Spaced Numbers Between 0 and 5

```python
arr = np.linspace(0, 5, 20)
print(arr)
```

## 3. Extract Center 3x3 from 5x5

```python
arr = np.random.randint(10, 100, (5, 5))
center = arr[1:4, 1:4]
print("5x5:\n", arr)
print("center 3x3:\n", center)
```

## 4. Transpose 2x2

```python
arr = np.random.randint(10, 100, (2, 2))
print("original:\n", arr)
print("transposed:\n", arr.T)
```

## 5. Reshape (2,6) → (3,4) and (4,3)

```python
arr = np.random.randint(10, 100, (2, 6))
print("original:\n", arr)
print("(3,4):\n", arr.reshape(3, 4))
print("(4,3):\n", arr.reshape(4, 3))
```

## 6. Flatten and Reshape Back

```python
arr = np.random.rand(2, 2)
flat = arr.flatten()
restored = flat.reshape(arr.shape)
print("flat:", flat)
print("restored:\n", restored)
```

## 7. Replace Negatives with 0

```python
arr = np.random.randint(-10, 50, (4, 4))
arr[arr < 0] = 0
print(arr)
```

## 8. Count Values Greater Than Mean

```python
arr = np.random.randint(10, 50, (4, 4))
count = np.sum(arr > arr.mean())
print("count > mean:", count)
```

## 9. Replace Even Numbers with Their Negatives

```python
arr = np.random.randint(10, 50, size=10)
arr[arr % 2 == 0] *= -1
print(arr)
```

## 10. Row Normalization (Sum = 1)

```python
arr = np.random.randint(1, 10, (4, 4))
normalized = arr / arr.sum(axis=1, keepdims=True)
print("normalized:\n", normalized)
print("row sums:", normalized.sum(axis=1))
```

## 11. Set Diagonal to 0

```python
arr = np.random.randint(1, 10, (5, 5))
np.fill_diagonal(arr, 0)
print(arr)
```

## 12. Missing Value Imputation (Column Mean)

```python
arr = np.random.rand(10, 4) * 100
mask = np.random.choice([True, False], arr.shape, p=[0.1, 0.9])
arr[mask] = np.nan

col_means = np.nanmean(arr, axis=0)
inds = np.where(np.isnan(arr))
arr[inds] = np.take(col_means, inds[1])

print("no NaNs remaining:", not np.isnan(arr).any())
print(arr)
```

## 13. Feature Standardization (Z-Score)

```python
arr = np.random.randint(1, 100, (5, 4)).astype(float)
standardized = (arr - arr.mean(axis=0)) / arr.std(axis=0)
print(standardized)
```

## 14. One-Hot Encoding

```python
colors = np.random.choice(['red', 'blue', 'green'], size=10)
unique_colors = np.unique(colors)
one_hot = (colors[:, None] == unique_colors[None, :]).astype(int)

print("labels:", colors)
print("classes:", unique_colors)
print("one-hot:\n", one_hot)
```

## 15. Moving Average

```python
arr = np.arange(10)
window = 3
moving_avg = np.convolve(arr, np.ones(window) / window, mode='valid')
print(moving_avg)
```

## 16. Manual Matrix Multiplication

```python
A = np.random.randint(1, 5, (2, 3))
B = np.random.randint(1, 5, (3, 2))

result = np.zeros((A.shape[0], B.shape[1]))
for i in range(A.shape[0]):
    for j in range(B.shape[1]):
        for k in range(A.shape[1]):
            result[i, j] += A[i, k] * B[k, j]

print("manual:\n", result)
print("np.dot check:\n", np.dot(A, B))
print("match:", np.allclose(result, np.dot(A, B)))
```

## 17. Simple Image Blur (3x3 Mean Filter)

```python
image = np.random.randint(0, 256, (10, 10))
blurred = np.zeros_like(image, dtype=float)

for i in range(1, image.shape[0] - 1):
    for j in range(1, image.shape[1] - 1):
        blurred[i, j] = np.mean(image[i-1:i+2, j-1:j+2])

print("original:\n", image)
print("blurred:\n", blurred)
```

## 18. Weather Data Simulation

```python
temps = np.random.randint(-5, 40, 365)
rolling_avg = np.convolve(temps, np.ones(7) / 7, mode='valid')

print("hottest day:", np.max(temps))
print("coldest day:", np.min(temps))
print("7-day rolling avg (first 10):", rolling_avg[:10])
```

## 19. Recommendation Matrix — Fill NaN with Item Average

```python
ratings = np.random.randint(1, 6, (5, 5)).astype(float)
mask = np.random.choice([True, False], ratings.shape, p=[0.2, 0.8])
ratings[mask] = np.nan

col_means = np.nanmean(ratings, axis=0)
inds = np.where(np.isnan(ratings))
ratings[inds] = np.take(col_means, inds[1])

print(ratings)
```

## 20. Stock Price Analysis

```python
prices = np.random.randint(100, 200, 30)
returns = np.diff(prices)
best_day = np.argmax(returns) + 1

print("prices:", prices)
print("daily returns:", returns)
print("best buy day:", best_day)
print("max single-day gain:", returns.max())
```
